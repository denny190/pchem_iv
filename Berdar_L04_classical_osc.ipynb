Classical harmonic oscillator: plot the probability distribution of observing a certain displacement at  (1) fixed energy for various values of E/h*nu; (2) fixed temperature for various values of T/Theta_v; Hint: numerically integrate the configuration integral in one dimension

Potential energy for a 1D harmonic oscillator:

$U(x) = \frac{1}{2}kx^2$


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants => set as arbitrary
k = 1       # Force constant
hbar = 1    # Reduced Planck's constant
nu = 1      # Frequency
kB = 1      # Boltzmann constant

# For fixed energy:
E_values = [0.5, 1, 2, 3]    # Sample energy values
x_range = np.linspace(-3, 3, 1000)
for E in E_values:
    plt.plot(x_range, np.where(np.abs(0.5*k*x_range**2 - E) < 0.05, 1, 0), label=f'E/hbar*nu = {E}')

plt.title('Probability distribution for fixed energy values')
plt.xlabel('Displacement (x)')
plt.ylabel('Probability')
plt.legend()
plt.show()

# For fixed temperature:
T_values = [0.5, 1, 2, 3]
x_range = np.linspace(-3, 3, 1000)
for T in T_values:
    beta = 1 / (kB * T)
    P = np.exp(-beta * 0.5 * k * x_range**2)
    normalization = np.trapz(P, x_range)    # Numerical integration
    plt.plot(x_range, P/normalization, label=f'T/Theta_v = {T}')

plt.title('Probability distribution for fixed temperature values')
plt.xlabel('Displacement (x)')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
